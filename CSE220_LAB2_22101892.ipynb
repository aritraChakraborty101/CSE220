```python
mul_arr = [[""] * 10, [""] * 10]
first_start = 0
second_start = 0

first_inp = input("Enter your first line ")
second_inp = input("Enter your second letter ")


# Checking for invalid input and pushing strings in an array
if len(first_inp) > 10 and len(second_inp) > 10:
    print("Invalid input size")
else:
    for i in range(len(first_inp)):
        mul_arr[0][i] = first_inp[i]
    for i in range(len(second_inp)):
        mul_arr[1][i] = second_inp[i]

# Excluding empty strings from couting as the size
size_1 = 0
size_2 = 0
for i in range(len(mul_arr[0])):
  if mul_arr[0][1] != "":
    size_1 +=1
  else:
    break
for i in range(len(mul_arr[1])):
  if mul_arr[1][1] != "":
    size_2 +=1
  else:
    break

# Finding starting index
for i in range(len(mul_arr[0])):
  if mul_arr[0][i] == '':
    continue
  else:
    if 64 < ord(mul_arr[0][i]) < 91:
        first_start = i
        break
    else:
        pass
for i in range(len(mul_arr[1])):
  if mul_arr[0][i] == '':
    continue
  else:
    if 64 < ord(mul_arr[1][i]) < 91:
        second_start = i
        break
    else:
        pass
        

inc_first = first_start
inc_second = second_start
print("Your input is valid. Press any key and then press enter to continue !!! ")
print(f"Top board start character: {mul_arr[0][first_start]}")
print(f"Top board start index: {first_start}")
print(f"Bottom board start: {mul_arr[1][second_start]}")
print(f"Bottom board start: {second_start}")

flag = True
#user_inp = input(" Your input is valid. Press any key and then press enter to continue !!! ")
while flag:
  user_inp = input("Press any key and then press enter to continue !!! ")
  if user_inp == "q" or user_inp == "Q":
    flag = False
    print(f"Top board start character: {mul_arr[0][first_start]}")
    print(f"Top board start index: {first_start}")
    print(f"Bottom board start: {mul_arr[1][second_start]}")
    print(f"Bottom board start: {second_start}")
    break
  else:
  
    # For first line 
    for i in range(len(mul_arr[0])):
      print(mul_arr[0][(inc_first-i)%size_1],end = "")
    inc_first-=1
    print()
    # For second line
    
    for i in range(len(mul_arr[1])):
      print(mul_arr[1][(inc_second+i)%size_2], end="")
    inc_second +=1
    print()
```
