```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
  def __init__(self, a):
    self.head = Node(a[0], None, None)
    self.head.next = self.head
    self.head.prev = self.head
    self.tail = self.head
    for i in range(1,len(a)):
      node = Node(a[i], self.head,self.tail)
      self.tail.next = node
      self.tail = self.tail.next
      self.head.prev = self.tail
  
  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    count = 0
    new = self.head
    while new.next is not self.head:
      count += 1
      new = new.next
    new = new.next
    return count + 1
  
  # prints the elements in the list
  def forwardprint(self):
    new = self.head
    while new.next != self.head:
      print(new.element, end=",")
      new = new.next
    print(new.element)
    if new.next == self.head:
        return new.next


  # prints the elements in the list backward
  def backwardprint(self):
    new = self.head
    while new.prev != self.head:
      new = new.prev
      print(new.element, end=",")
      
    print(new.prev.element)

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    length = self.countNode()
    node = self.head
    count = 0
    if idx >= length:
      return None # invalid
    else:
      while node.next != self.head:
        if count == idx:
          return node
        count += 1
        node = node.next
      return node


  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    length = self.countNode()
    new = self.head
    idx = elem
    count = -1
    while new.next is not self.head:
      if elem != new.element:
        new =  new.next
        count +=1
      elif elem == new.element:
        count += 1
        break
    if new.element == elem:
      count+=1
      return count
    return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    count = self.countNode()
    if idx > count:
      print("Invalid index.")
    else:
      new = Node(elem, None, None)
      if idx == 0:
        temp = self.head
        head = new
        tail = self.head.prev
        tail.next = head
        head.prev = tail
        head.next = temp
        head.next.prev = head
        self.head = head

      elif idx > 0 and idx < count:
        temp_1 = self.nodeAt(idx)
        temp_2 = temp_1.prev
        new.prev = temp_2
        temp_2.next = new
        new.next = temp_1
        temp_1.prev = new
      else:
        head = self.head
        temp = self.nodeAt(count-1)
        temp.next = new
        new.prev = temp
        new.next = head
        head.prev = new
  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    count = self.countNode()
    if idx >= count:
      return None
    else:
      if idx == 0:
        self.head.prev.next = self.head.next
        self.head.next.prev = self.head.prev
        self.head = self.head.next
        return str(self.head.element)
      else:
        removed = self.nodeAt(idx)
        removed.prev.next = removed.next
        removed.next.prev = removed.prev
        return str(removed.element)
```
